# Morphological opeations

## Erosion 

```python
import cv2
import numpy
import pandas
from matplotlib import pyplot as plt
```

```python
gray = cv2.imread("C:\\Users\\rishi\\OneDrive\\Desktop\\group icon.jfif")
```

```python
plt.imshow(gray)
```

```python
for i in range(0,3):
    eroded = cv2.erode(gray.copy(), None, iterations = i + 1)
    cv2.imshow("Eroded {} times".format(i + 1), eroded)
    cv2.waitKey(0)
```

## Dilation

```python
for i in range(0, 3):
    dilated = cv2.dilate(gray.copy(), None, iterations=i + 1)
    cv2.imshow("Dilated {} times".format(i + 1), dilated)
    cv2.waitKey(0)
```

## Opening

```python
kernelSizes = [(3, 3), (5, 5), (7, 7)]
for kernelSize in kernelSizes:
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernelSize)
    opening = cv2.morphologyEx(gray, cv2.MORPH_OPEN, kernel)
    cv2.imshow("Opening: ({}, {})".format(kernelSize[0], kernelSize[1]), opening)
    cv2.waitKey(0)
```

## Closing

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernelSize)
for kernelSize in kernelSizes:
    closing = cv2.morphologyEx(gray, cv2.MORPH_CLOSE, kernel)
    cv2.imshow("Closing: ({}, {})".format(kernelSize[0], kernelSize[1]), closing)
    cv2.waitKey(0)
```

## Morphological Gradient

```python
for kernelSize in kernelSizes:
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernelSize)
    gradient = cv2.morphologyEx(gray, cv2.MORPH_GRADIENT, kernel)
    cv2.imshow("Gradient: ({}, {})".format(kernelSize[0], kernelSize[1]), gradient)
    cv2.waitKey(0)
```

## top hat (used to identify bright regions from a dark background)

```python
image = cv2.imread("C:\\Users\\rishi\\OneDrive\\Desktop\\vantablack.jpg")
```

```python
gray2 = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
rectKernel = cv2.getStructuringElement(cv2.MORPH_RECT, (13, 5))
blackhat = cv2.morphologyEx(gray2, cv2.MORPH_BLACKHAT, rectKernel)
```

```python
tophat = cv2.morphologyEx(gray2, cv2.MORPH_TOPHAT, rectKernel)
```

```python
cv2.imshow("Original", image)
cv2.imshow("Blackhat", blackhat)
cv2.imshow("Tophat", tophat)
cv2.waitKey(0)
```
