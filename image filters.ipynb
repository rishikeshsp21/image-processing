# Image Filters

```python
import cv2
import numpy
import pandas
from matplotlib import pyplot as plt
```

## median blur

### Useful in removing salt and pepper noise

```python
salt = cv2.imread("salty.jfif")
plt.imshow(salt)
```

```python
nosalt = cv2.medianBlur(salt, 5)
plt.imshow(nosalt)
```

## laplacian Filter

### Useful in finding image edges by highlighting gray level discontinuities

```python
edge1 = cv2.imread("giza.jpg")
plt.imshow(edge1)
```

```python
#blurring the image and converting it to a grayscale image
edge2 =  cv2.GaussianBlur(edge1, (21,21), 0)
blurgray = cv2.cvtColor(edge2, cv2.COLOR_BGR2GRAY)
dst = cv2.Laplacian(blurgray, cv2.CV_16S, ksize=3)
plt.imshow(dst)
```

## Gaussian Filter

### Gaussian kernel weights pixels at its center much more strongly than its boundaries

```python
plt.imshow(edge1)
gray2 = cv2.cvtColor(edge1, cv2.COLOR_BGR2GRAY)
plt.imshow(gray2)
```

```python
gblur = cv2.GaussianBlur(edge1, (21,21), 0)
plt.imshow(gblur)
```
